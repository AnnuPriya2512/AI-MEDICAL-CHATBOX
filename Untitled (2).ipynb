```python
import speech_recognition as sr
import time

recognizer = sr.Recognizer()
mic = sr.Microphone()

conversation = []
silence_start = None
silence_limit = 20   # stop if silence for 20 seconds

print("Listening... (Press Ctrl+C to end manually or stay silent for 20s)")

try:
    with mic as source:
        recognizer.adjust_for_ambient_noise(source)
        while True:
            try:
                print("Doctor/Patient, please speak...")
                # Listen for up to 5 seconds at a time
                audio = recognizer.listen(source, timeout=None, phrase_time_limit=10)

                text = recognizer.recognize_google(audio)
                print("Heard:", text)
                conversation.append(text)

                # reset silence timer since we got speech
                silence_start = None  

            except sr.UnknownValueError:
                print("Could not understand audio.")
                # start silence timer if not already started
                if silence_start is None:
                    silence_start = time.time()
                else:
                    if time.time() - silence_start > silence_limit:
                        print("\n🛑 Auto-stopped due to long silence.")
                        break

            except sr.RequestError:
                print("API error or network issue.")

except KeyboardInterrupt:
    print("\n🛑 Conversation ended by doctor (Ctrl+C).")

# Save the full conversation
if conversation:
    with open("conversation.txt", "w") as f:
        f.write("\n".join(conversation))
    print("✅ Full conversation saved to 'conversation.txt'")
else:
    print("⚠️ No conversation recorded.")
```

```python
import spacy
import re
import json

# Load SciSpaCy medical model
nlp = spacy.load("en_core_sci_sm")

def extract_medical_info(conversation_text):
    doc = nlp(conversation_text)

    symptoms = []
    duration = []
    medications = []
    diagnosis = []

    # Extract entities using SciSpaCy
    for ent in doc.ents:
        if ent.label_.lower() in ["symptom", "disease", "condition"]:
            symptoms.append(ent.text)
        elif "day" in ent.text.lower() or "week" in ent.text.lower():
            duration.append(ent.text)

    # Regex-based symptom detection
    regex_symptoms = re.findall(
        r"\b(headache|pain|nausea|fever|blurred vision|sensitivity to light|throbbing)\b",
        conversation_text,
        re.IGNORECASE
    )
    symptoms.extend(regex_symptoms)

    # Possible diagnosis
    if "migraine" in conversation_text.lower():
        diagnosis.append("migraine")

    # Possible medication
    if "prescribe" in conversation_text.lower():
        medications.append("pain medication")
        medications.append("anti-nausea medication")

    # Create structured data
    extracted_data = {
        "symptoms": list(set(symptoms)),
        "duration": list(set(duration)),
        "diagnosis": list(set(diagnosis)),
        "medications": list(set(medications))
    }

    return extracted_data


if __name__ == "__main__":
    # Read conversation from file
    with open("conversation.txt", "r", encoding="utf-8") as f:
        conversation = f.read()

    # Extract data
    data = extract_medical_info(conversation)

    # Save extracted data into JSON
    with open("extracted_medical_data.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)

    print("✅ Extracted medical info saved to extracted_medical_data.json")
```

```python
# Run this whole cell in Jupyter (assumes conversation.txt exists in the same folder).
# The script will try to use spaCy/sciSpacy if available; if not it uses regex-based extraction.

import re
import json
import csv
from collections import defaultdict

# Try to import spaCy (sciSpacy model if present) for better sentence splitting and NER.
has_spacy = False
nlp = None
try:
    import spacy
    # Try SciSpaCy model first, then fallback to small web model
    try:
        nlp = spacy.load("en_core_sci_sm")
    except Exception:
        try:
            nlp = spacy.load("en_core_web_sm")
        except Exception:
            nlp = None
    if nlp is not None:
        has_spacy = True
except Exception:
    has_spacy = False


# --- Helper utilities -------------------------------------------------------

NUM_WORDS = {
    "one":1,"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7,"eight":8,"nine":9,"ten":10,
    "eleven":11,"twelve":12,"thirteen":13,"fourteen":14,"fifteen":15,"sixteen":16,"seventeen":17,
    "eighteen":18,"nineteen":19,"twenty":20
}

SYMPTOM_KEYWORDS = [
    "headache","nausea","vomit","vomiting","fever","chills","cough","sore throat","shortness of breath",
    "breathlessness","dizziness","weakness","fatigue","blurred vision","sensitivity to light","photophobia",
    "abdominal pain","chest pain","pain","throbbing","nauseous","nausea and vomiting","diarrhea","constipation",
    "vomit","loss of appetite","weight loss","weight gain","palpitations","sweating","itching","rash"
]
# Make regex-friendly unique sorted list
SYMPTOM_KEYWORDS = sorted(set(SYMPTOM_KEYWORDS), key=lambda s: -len(s))

DIAGNOSIS_KEYWORDS = ["migraine","infection","flu","covid","pneumonia","hypertension","diabetes","appendicitis","fracture"]
MED_NAMES = ["paracetamol","acetaminophen","ibuprofen","aspirin","amoxicillin","ondansetron","metoclopramide"]


def read_conversation(path="conversation.txt"):
    with open(path, "r", encoding="utf-8") as f:
        raw = f.read()
    return raw


def normalize_text(text):
    # remove "Doctor:" / "Patient:" labels (case-insensitive) and collapse spaces
    t = re.sub(r'(?mi)^\s*(doctor|patient)\s*:\s*', '', text)  # lines starting with Doctor:/Patient:
    t = re.sub(r'(?mi)\b(doctor|patient)\s*:\s*', '', t)  # inline labels if any
    t = t.replace("\r\n", "\n")
    t = re.sub(r'\s+\n', '\n', t)
    t = t.strip()
    return t


def find_sentences(text):
    if has_spacy:
        doc = nlp(text)
        return [sent.text.strip() for sent in doc.sents]
    else:
        # simple sentence split fallback
        sents = re.split(r'(?<=[.!?])\s+', text)
        return [s.strip() for s in sents if s.strip()]


def wordnum_to_int(word):
    w = word.lower().strip()
    return NUM_WORDS.get(w)


# --- Field extractors ------------------------------------------------------

def extract_age(text):
    # digits: "42 years old", "age 42", "I am 42"
    patterns = [
        r'\b(\d{1,3})\s*(?:years? old|yrs? old|y/o)\b',
        r'\bage\s*(?:is|:)?\s*(\d{1,3})\b',
        r'\bi am\s+(\d{1,3})\b'
    ]
    for p in patterns:
        m = re.search(p, text, re.IGNORECASE)
        if m:
            try:
                return int(m.group(1))
            except:
                pass
    # try word numbers like "forty two years" (limited)
    m = re.search(r'(\b(?:one|two|three|four|five|six|seven|eight|nine|ten|eleven|twelve|thirteen|fourteen|fifteen|sixteen|seventeen|eighteen|nineteen|twenty)\b)\s+years', text, re.IGNORECASE)
    if m:
        val = wordnum_to_int(m.group(1))
        if val: return val
    return None


def extract_weight(text):
    # kg or lbs
    m = re.search(r'(\d{1,3}(?:\.\d+)?)\s*(kg|kilograms|kgs)\b', text, re.IGNORECASE)
    if m:
        return {"value": float(m.group(1)), "unit":"kg"}
    m2 = re.search(r'(\d{1,3}(?:\.\d+)?)\s*(lb|lbs|pounds)\b', text, re.IGNORECASE)
    if m2:
        val = float(m2.group(1))
        return {"value": round(val * 0.453592, 2), "unit":"kg", "original": f"{val} lbs"}
    return None


def extract_durations(text):
    durations = []
    # patterns like "for the past 3 days", "since yesterday", "3 days ago", "for three days"
    p1 = re.compile(r'\b(?:for|since|past|over|about|around)\s+(?:the\s+)?(?:(\d+)|(' + '|'.join(NUM_WORDS.keys()) + r'))\s+(day|days|week|weeks|month|months|hour|hours)\b', re.I)
    for m in p1.finditer(text):
        num = m.group(1) or m.group(2)
        if num is not None:
            if num.isdigit():
                qty = int(num)
            else:
                qty = wordnum_to_int(num)
            unit = m.group(3)
            if qty:
                durations.append(f"{qty} {unit}")
    # patterns like "for 3 days", "3 days ago"
    p2 = re.compile(r'\b(\d+)\s+(days?|weeks?|months?|hours?)\b', re.I)
    for m in p2.finditer(text):
        durations.append(f"{m.group(1)} {m.group(2)}")
    # "since yesterday", "yesterday", "today", "last week"
    p3 = re.compile(r'\b(since\s+yesterday|yesterday|today|last\s+week|last\s+month)\b', re.I)
    for m in p3.finditer(text):
        durations.append(m.group(0))
    return list(dict.fromkeys(durations))  # unique preserve order


def detect_negation(snippet, symptom):
    # simple heuristic: if "no/denies/not/without" appears within 6 tokens before symptom -> negated
    # snippet is the nearby text; symptom is the matched word
    # we will search preceding up to 6 words
    lower = snippet.lower()
    # find symptom position
    idx = lower.find(symptom.lower())
    window = lower[:idx] if idx>=0 else lower
    # check patterns
    if re.search(r'\b(no|not|never|denies|denied|without)\b', window[-100:]):
        return True
    # also patterns like "denies having headache" or "no fever"
    return False


def find_nearby_info(full_text, match_start, match_end, window_chars=80):
    a = max(0, match_start - window_chars)
    b = min(len(full_text), match_end + window_chars)
    return full_text[a:b]


def extract_symptoms(text):
    lower = text.lower()
    found = []
    for kw in SYMPTOM_KEYWORDS:
        # word boundary search
        for m in re.finditer(r'\b' + re.escape(kw) + r'\b', lower):
            start, end = m.start(), m.end()
            snippet = find_nearby_info(lower, start, end, window_chars=80)
            neg = detect_negation(snippet, kw)
            # severity
            sev_m = re.search(r'\b(mild|moderate|severe|throbbing|sharp|dull|constant|intermittent|slight|intense)\b', snippet)
            severity = sev_m.group(1) if sev_m else None
            # laterality
            lat_m = re.search(r'\b(left|right|bilateral|left-sided|right-sided)\b', snippet)
            laterality = lat_m.group(1) if lat_m else None
            # duration near symptom
            dur_m = re.search(r'(for\s+(?:\d+\s+)?(?:days?|weeks?|months?|hours?)|since\s+\w+|past\s+\d+\s+\w+)', snippet)
            duration = dur_m.group(0) if dur_m else None
            found.append({
                "symptom": kw,
                "negated": bool(neg),
                "severity": severity,
                "laterality": laterality,
                "duration": duration,
                "context": snippet.strip()
            })
    # deduplicate and merge: prefer non-negated entries, accumulate descriptors
    agg = {}
    for item in found:
        key = item["symptom"]
        if key not in agg:
            agg[key] = item
        else:
            # merge severity / laterality / duration if missing
            if not agg[key]["severity"] and item["severity"]:
                agg[key]["severity"] = item["severity"]
            if not agg[key]["laterality"] and item["laterality"]:
                agg[key]["laterality"] = item["laterality"]
            if not agg[key]["duration"] and item["duration"]:
                agg[key]["duration"] = item["duration"]
            # if one occurrence is negated and another not, prefer not-negated
            if agg[key]["negated"] and not item["negated"]:
                agg[key]["negated"] = False
    return list(agg.values())


def extract_possible_diagnoses(text):
    found = []
    lower = text.lower()
    for kw in DIAGNOSIS_KEYWORDS:
        if re.search(r'\b' + re.escape(kw) + r'\b', lower):
            # check if it is mentioned in context like "sounds like" "could be" "diagnosis" "likely"
            # capture phrase around match
            m = re.search(r'(.{0,40}\b' + re.escape(kw) + r'\b.{0,40})', lower)
            ctx = m.group(0) if m else kw
            found.append({"diagnosis": kw, "context": ctx.strip()})
    # also phrases like "it could be a migraine" - handled by keyword presence
    return found


def extract_medications_and_prescription(text):
    meds = []
    lower = text.lower()
    # explicit medication names
    for name in MED_NAMES:
        if re.search(r'\b' + re.escape(name) + r'\b', lower):
            meds.append({"name": name})
    # if doctor said "prescribe" or "I'll prescribe some medication", record generic
    if re.search(r'\bprescrib(e|ed|ing)\b', lower):
        # try to find what is prescribed
        m = re.search(r'prescrib(?:e|ed|ing)\s+(?:some\s+)?([a-z\s\-]+?)(?:\.|,|for|to|$)', lower)
        if m:
            meds.append({"name": m.group(1).strip()})
        else:
            # add generics often used
            if "nausea" in lower or "nauseous" in lower:
                meds.append({"name": "anti-nausea medication (generic)"})
            if "pain" in lower or "headache" in lower:
                meds.append({"name": "pain medication (analgesic, generic)"})
    return meds


# --- Main pipeline ---------------------------------------------------------

def extract_medical_info_from_file(path="conversation.txt"):
    raw = read_conversation(path)
    normalized = normalize_text(raw)
    sentences = find_sentences(normalized)

    age = extract_age(normalized)
    weight = extract_weight(normalized)
    durations = extract_durations(normalized)
    symptoms = extract_symptoms(normalized)
    diagnoses = extract_possible_diagnoses(normalized)
    medications = extract_medications_and_prescription(normalized)

    # If spaCy available, also collect its entities for debug/extra
    spaCy_entities = []
    if has_spacy and nlp:
        doc = nlp(normalized)
        for ent in doc.ents:
            spaCy_entities.append({"text": ent.text, "label": ent.label_})

    result = {
        "age": age,
        "weight": weight,
        "durations_found": durations,
        "symptoms": symptoms,
        "possible_diagnoses": diagnoses,
        "medications": medications,
        "spaCy_entities": spaCy_entities,
        "raw_conversation": normalized
    }
    return result


def save_outputs(result, json_path="extracted_medical_data.json", csv_path="extracted_medical_data.csv"):
    # Save JSON
    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(result, f, indent=2, ensure_ascii=False)

    # Also save a CSV summary: one row per conversation (flatten key fields)
    row = {
        "age": result.get("age"),
        "weight_kg": (result.get("weight") or {}).get("value") if result.get("weight") else None,
        "weight_original": (result.get("weight") or {}).get("original") if result.get("weight") else None,
        "symptoms": "; ".join([s["symptom"] + (" (neg)" if s["negated"] else "") for s in result["symptoms"]]) if result["symptoms"] else "",
        "possible_diagnoses": "; ".join([d["diagnosis"] for d in result["possible_diagnoses"]]) if result["possible_diagnoses"] else "",
        "medications": "; ".join([m["name"] for m in result["medications"]]) if result["medications"] else ""
    }
    # write header if file doesn't exist, else append
    try:
        with open(csv_path, "x", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=row.keys())
            writer.writeheader()
            writer.writerow(row)
    except FileExistsError:
        with open(csv_path, "a", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=row.keys())
            writer.writerow(row)


# --- Execute pipeline ------------------------------------------------------

if __name__ == "__main__":
    result = extract_medical_info_from_file("conversation.txt")
    save_outputs(result)
    print("✅ Extraction complete.")
    print("Saved: extracted_medical_data.json  and extracted_medical_data.csv")
    # print a short readable summary
    print("\n--- Summary ---")
    print("Age:", result['age'])
    print("Weight:", result['weight'])
    print("Symptoms found:")
    for s in result['symptoms']:
        print(" -", s['symptom'], "| negated:" , s['negated'], "| severity:", s['severity'], "| laterality:", s['laterality'], "| duration:", s['duration'])
    print("Possible diagnoses:", [d['diagnosis'] for d in result['possible_diagnoses']])
    print("Medications/prescription:", [m['name'] for m in result['medications']])
    if has_spacy:
        print("\nNote: spaCy model detected; additional entities saved in spaCy_entities.")
    else:
        print("\nNote: spaCy not available; regex-based extraction used. To improve results, install SciSpaCy (optional).")
```

```python
import re

def read_conversation(file_path="conversation.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

def extract_medical_parameters(text):
    extracted = {}

    # Extract Age
    age_match = re.search(r'(\d{1,3})\s*(?:years? old|yrs? old|y/o|years)', text, re.IGNORECASE)
    if age_match:
        extracted["Age"] = age_match.group(1)

    # Extract Weight
    weight_match = re.search(r'(\d{1,3})\s*(kg|kgs|kilograms|pounds|lbs)', text, re.IGNORECASE)
    if weight_match:
        extracted["Weight"] = weight_match.group(1) + " " + weight_match.group(2)

    # Extract Symptoms (simple keyword-based for Stage 1)
    symptoms_keywords = [
        "headache", "nausea", "vomiting", "fever", "cough", 
        "chest pain", "dizziness", "fatigue", "sore throat",
        "blurred vision", "sensitivity to light"
    ]
    symptoms_found = []
    for symptom in symptoms_keywords:
        if re.search(r"\b" + re.escape(symptom) + r"\b", text, re.IGNORECASE):
            symptoms_found.append(symptom)
    if symptoms_found:
        extracted["Symptoms"] = symptoms_found

    # Extract Duration
    duration_match = re.search(r'(\d+)\s*(day|days|week|weeks|month|months)', text, re.IGNORECASE)
    if duration_match:
        extracted["Duration"] = duration_match.group(0)

    return extracted

def save_extracted_parameters(data, file_path="medical_parameters.txt"):
    with open(file_path, "w", encoding="utf-8") as f:
        for key, value in data.items():
            if isinstance(value, list):
                f.write(f"{key}: {', '.join(value)}\n")
            else:
                f.write(f"{key}: {value}\n")

# --- Run Stage 1 ---
conversation_text = read_conversation("conversation.txt")
medical_data = extract_medical_parameters(conversation_text)
save_extracted_parameters(medical_data)

print("✅ Medical parameters extracted and saved to medical_parameters.txt")
print(medical_data)
```

```python
# Copy this entire cell into Jupyter and run.
# It will:
# - read medical_parameters.txt (preferred) or fallback to conversation.txt (if present)
# - load the disease CSV at /mnt/data/prostate_diseases_symptoms_age_comorbidities.csv
# - compute match scores and save top predictions to predicted_diseases.json and predicted_diseases.txt

import pandas as pd
import re
import json
from pathlib import Path

CSV_PATH = Path("prostate_diseases_symptoms_age_comorbidities.csv")
MED_PARAMS_PATH = Path("medical_parameters.txt")
CONV_PATH = Path("conversation.txt")
OUT_JSON = Path("predicted_diseases.json")
OUT_TXT = Path("predicted_diseases.txt")

# ----------------- helpers -----------------
def read_med_params(path):
    txt = path.read_text(encoding="utf-8")
    data = {}
    for line in txt.splitlines():
        if ':' in line:
            k,v = line.split(':',1)
            data[k.strip().lower()] = v.strip()
    # normalize
    age = None
    if 'age' in data:
        m = re.search(r'(\d{1,3})', data['age'])
        if m: age = int(m.group(1))
    else:
        m = re.search(r'(\d{1,3})\s*(?:years?|yrs?|y/o)', txt, re.IGNORECASE)
        if m: age = int(m.group(1))
    symptoms = []
    if 'symptoms' in data:
        symptoms = [s.strip().lower() for s in re.split(r'[;,]', data['symptoms']) if s.strip()]
    # detect common comorbid keywords inside the full text
    comorb_keywords = ['diabetes','hypertension','high blood pressure','bp','asthma','obesity','cardiac','heart','smoker']
    comorbs = [kw for kw in comorb_keywords if kw in txt.lower()]
    return {"age": age, "symptoms": symptoms, "comorbidities": comorbs, "raw": txt}

def extract_from_conversation(path):
    txt = path.read_text(encoding="utf-8").lower()
    # Age
    age = None
    m = re.search(r'(\d{1,3})\s*(?:years? old|y/o|yrs? old|years)', txt)
    if m: age = int(m.group(1))
    # Symptoms: find any disease symptom keywords present in CSV later (we'll detect after loading CSV)
    # Comorbidities detection (keywords)
    comorb_keywords = ['diabetes','hypertension','high blood pressure','bp','asthma','obesity','cardiac','heart','smoker']
    comorbs = [kw for kw in comorb_keywords if kw in txt]
    return {"age": age, "symptoms": [], "comorbidities": comorbs, "raw": txt}

def parse_disease_row(r):
    s = (r.get('symptoms') or "") 
    items = [x.strip().lower() for x in re.split(r'[;,]', s) if x.strip()]
    age_txt = r.get('typical_age_group') or ""
    nums = re.findall(r'(\d{1,3})', age_txt)
    age_range = None
    if len(nums) >= 2:
        age_range = (int(nums[0]), int(nums[1]))
    elif len(nums) == 1:
        v = int(nums[0]); age_range = (max(0, v-10), v+10)
    com = r.get('related_comorbidities') or ""
    com_list = [x.strip().lower() for x in re.split(r'[;,]', com) if x.strip()]
    return {"disease": r.get('disease'), "symptoms": items, "age_range": age_range, "comorbidities": com_list, "age_text": age_txt}

def symptom_overlap_score(patient_symptoms, disease_symptoms):
    if not disease_symptoms:
        return 0.0
    pat = [s.lower() for s in patient_symptoms]
    dis = [s.lower() for s in disease_symptoms]
    match_count = 0
    for ps in pat:
        for ds in dis:
            if ps == ds or ps in ds or ds in ps:
                match_count += 1
                break
    return match_count / max(1, len(dis))

def age_score(patient_age, age_range):
    if patient_age is None or age_range is None:
        return 0.0
    low, high = age_range
    if low <= patient_age <= high: return 1.0
    if patient_age >= low-10 and patient_age <= high+10: return 0.6
    return 0.0

def comorbidity_score(patient_comorbs, disease_comorbs):
    if not patient_comorbs or not disease_comorbs: return 0.0
    pc = set(patient_comorbs); dc = set(disease_comorbs)
    inter = pc.intersection(dc)
    return len(inter) / max(1, len(dc))

# -------------- main flow -------------------
if not CSV_PATH.exists():
    print("ERROR: disease CSV not found at:", CSV_PATH)
else:
    # Load disease CSV and parse rows
    df = pd.read_csv(CSV_PATH)
    disease_list = [parse_disease_row(df.loc[i].to_dict()) for i in range(len(df))]

    # Load patient data (prefer medical_parameters.txt)
    if MED_PARAMS_PATH.exists():
        patient = read_med_params(MED_PARAMS_PATH)
    elif CONV_PATH.exists():
        patient = extract_from_conversation(CONV_PATH)
        # if we pulled symptoms empty above, populate symptom list by scanning for disease symptom keywords
        # gather all symptom keywords from dataset
        keyword_set = set(sum([d['symptoms'] for d in disease_list], []))
        found = []
        for kw in keyword_set:
            # simple substring match in raw conversation
            if kw and kw in patient['raw'].lower():
                found.append(kw)
        patient['symptoms'] = sorted(set(found))
    else:
        print("ERROR: Neither 'medical_parameters.txt' nor 'conversation.txt' found in notebook folder.")
        print("Please place medical_parameters.txt (the extracted symptoms/age) in the same folder and re-run this cell.")
        raise FileNotFoundError("No input file available")

    # Ensure patient symptoms lowercase and unique
    patient['symptoms'] = [s.lower().strip() for s in patient.get('symptoms', []) if s and s.strip()]
    patient['comorbidities'] = [c.lower().strip() for c in patient.get('comorbidities', []) if c and c.strip()]

    # Compute match for each disease
    results = []
    for d in disease_list:
        s_score = symptom_overlap_score(patient['symptoms'], d['symptoms'])
        a_score = age_score(patient['age'], d['age_range'])
        c_score = comorbidity_score(patient['comorbidities'], d['comorbidities'])
        final = 0.6*s_score + 0.25*a_score + 0.15*c_score
        results.append({
            "disease": d['disease'],
            "final_score": round(final, 4),
            "symptom_score": round(s_score,4),
            "age_score": round(a_score,4),
            "comorb_score": round(c_score,4),
            "disease_symptoms": d['symptoms'],
            "disease_age_text": d['age_text']
        })

    results_sorted = sorted(results, key=lambda x: x['final_score'], reverse=True)
    top = results_sorted[:10]

    # Save outputs
    OUT_JSON.write_text(json.dumps({"patient": patient, "top_matches": top}, indent=2, ensure_ascii=False), encoding="utf-8")
    with OUT_TXT.open("w", encoding="utf-8") as f:
        f.write("Top predicted diseases (ranked):\n\n")
        for r in top:
            f.write(f"- {r['disease']}\n  final_score: {r['final_score']}\n  symptom_score: {r['symptom_score']}\n  age_score: {r['age_score']}\n  comorb_score: {r['comorb_score']}\n  disease_age_group_text: {r['disease_age_text']}\n  disease_symptoms: {', '.join(r['disease_symptoms'])}\n\n")

    # Print summary
    print("Patient data used for matching:")
    print(json.dumps(patient, indent=2))
    print("\nTop predicted diseases:")
    for r in top[:5]:
        print(f"{r['disease']} (score {r['final_score']})")

    print("\nSaved ->", OUT_JSON.resolve())
    print("Saved ->", OUT_TXT.resolve())
```

```python
# --- Vector Similarity between conversation symptoms and disease dataset ---
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Paths to your data (CSV files)
disease_file = "prostate_diseases_symptoms_age_comorbidities.csv"
conversation_file = "extracted_medical_data.csv"

# --- Load data from CSV ---
disease_df = pd.read_csv(disease_file)
conv_df = pd.read_csv(conversation_file)

# --- Combine relevant disease info for embedding ---
disease_df['combined_text'] = (
    disease_df['disease'].fillna('') + ' ' +
    disease_df['symptoms'].fillna('') + ' ' +
    disease_df['related_comorbidities'].fillna('')
)

# --- Combine extracted medical text from conversation ---
conv_text = ' '.join(conv_df.astype(str).fillna('').values.flatten())

# --- Load a pre-trained model for semantic similarity ---
model = SentenceTransformer('all-MiniLM-L6-v2')

# --- Encode the texts into dense vectors ---
disease_embeddings = model.encode(disease_df['combined_text'].tolist(), convert_to_tensor=True)
conv_embedding = model.encode([conv_text], convert_to_tensor=True)

# --- Compute cosine similarity (semantic similarity) ---
scores = cosine_similarity(conv_embedding, disease_embeddings).flatten()

# --- Add scores to dataframe ---
disease_df['vector_similarity_score'] = scores

# --- Sort and get top 10 matches ---
top_matches = disease_df.sort_values(by='vector_similarity_score', ascending=False).head(10)

# --- Display results ---
print("\nTop 10 diseases (by semantic vector similarity):")
print(top_matches[['disease', 'vector_similarity_score', 'symptoms', 'related_comorbidities']])

# --- Save results ---
top_matches.to_excel("vector_similarity_results.xlsx", index=False)
print("\n✅ Results saved as 'vector_similarity_results.xlsx'")
```

```python
# ==========================================
# Use DOCX conversations + disease CSV
# to extract medical terms & predict disease
# for EACH conversation in the DOCX file
# ==========================================

import re
from pathlib import Path

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# You may need this once:
# !pip install python-docx

from docx import Document

# ---------- FILE PATHS (edit names if needed) ----------
DOCX_FILE = Path("Patient Conversations 1 To 10.docx")
CSV_FILE  = Path("prostate_diseases_symptoms_age_comorbidities.csv")
OUTPUT_FILE = Path("conversation_disease_predictions.csv")

# ---------- SAFETY CHECKS ----------
if not DOCX_FILE.exists():
    raise FileNotFoundError(f"DOCX file not found: {DOCX_FILE}")
if not CSV_FILE.exists():
    raise FileNotFoundError(f"Disease CSV not found: {CSV_FILE}")

# ---------- 1. LOAD DISEASE DATA ----------
disease_df = pd.read_csv(CSV_FILE)

required_cols = ["disease", "symptoms", "typical_age_group", "related_comorbidities"]
for col in required_cols:
    if col not in disease_df.columns:
        raise ValueError(
            f"Column '{col}' missing in CSV. "
            f"Found columns: {list(disease_df.columns)}"
        )

# ---------- 2. PARSE AGE RANGE FROM CSV ----------
def parse_age_range(age_text):
    """
    Examples:
      '50-70', '40 to 60', 'above 60', 'below 40'
    Returns (low, high) or None.
    """
    if not isinstance(age_text, str):
        age_text = str(age_text)
    t = age_text.lower().strip()
    nums = re.findall(r'(\d{1,3})', t)

    if "above" in t or "over" in t:
        if nums:
            low = int(nums[0])
            return (low, 120)
    if "below" in t or "under" in t:
        if nums:
            high = int(nums[0])
            return (0, high)

    if len(nums) >= 2:
        low, high = int(nums[0]), int(nums[1])
        if low > high:
            low, high = high, low
        return (low, high)

    if len(nums) == 1:
        v = int(nums[0])
        return (max(0, v - 10), v + 10)

    return None

disease_df["parsed_age_range"] = disease_df["typical_age_group"].apply(parse_age_range)

# ---------- 3. BUILD SYMPTOM KEYWORDS FROM CSV ----------
def build_symptom_keywords(df):
    all_syms = []
    for s in df["symptoms"].fillna(""):
        parts = re.split(r"[;,]", str(s))
        for p in parts:
            p = p.strip().lower()
            if p:
                all_syms.append(p)
    # unique, preserve order
    seen = set()
    out = []
    for s in all_syms:
        if s not in seen:
            seen.add(s)
            out.append(s)
    # longer phrases first so we match "lower abdomen pain" before just "pain"
    out = sorted(out, key=len, reverse=True)
    return out

symptom_keywords = build_symptom_keywords(disease_df)

# ---------- 4. LOAD & SPLIT CONVERSATIONS FROM DOCX ----------
def load_docx_text(docx_path):
    doc = Document(docx_path)
    lines = [p.text for p in doc.paragraphs]
    return "\n".join(lines)

full_text = load_docx_text(DOCX_FILE)

def split_conversations_from_docx(text):
    """
    Splits by headings like 'Patient 3 (' or 'Patient 4 (' etc.
    Each 'Patient X (...)' block is treated as one conversation.
    """
    lines = text.splitlines()
    conversations = []
    current = []

    for line in lines:
        stripped = line.strip()
        # Start of a new conversation block
        if stripped.startswith("Patient ") and "Patient:" in stripped:
            # flush previous
            if current:
                conversations.append("\n".join(current).strip())
                current = []
        if stripped != "":
            current.append(stripped)

    # last block
    if current:
        conversations.append("\n".join(current).strip())

    # If for some reason nothing split, treat full text as one conversation
    if not conversations:
        conversations = [text.strip()]

    return conversations

conversations = split_conversations_from_docx(full_text)
print(f"🔍 Found {len(conversations)} conversation(s) in DOCX.\n")

# ---------- 5. HELPERS FOR PER-CONVERSATION PROCESSING ----------

def extract_age_from_text(text):
    text_low = text.lower()
    patterns = [
        r"(\d{1,3})\s*(?:years? old|yrs? old|y/o|years)",
        r"age\s*(?:is|:)?\s*(\d{1,3})",
        r"\b(\d{1,3})-year-old\b",
        r"\bi am\s+(\d{1,3})\b",
    ]
    for p in patterns:
        m = re.search(p, text_low)
        if m:
            try:
                return int(m.group(1))
            except:
                pass
    return None

def extract_symptoms_from_text(text, symptom_keywords):
    t = text.lower()
    found = []
    for kw in symptom_keywords:
        pattern = r"\b" + re.escape(kw) + r"\b"
        if re.search(pattern, t):
            found.append(kw)
    return sorted(set(found))

def age_score(patient_age, age_range):
    if patient_age is None or age_range is None:
        return 0.0
    low, high = age_range
    if low <= patient_age <= high:
        return 1.0
    if (patient_age >= low - 10) and (patient_age <= high + 10):
        return 0.5
    return 0.0

def compute_lexical_similarity(conv_text, disease_texts):
    texts = [conv_text] + disease_texts
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf = vectorizer.fit_transform(texts)
    conv_vec = tfidf[0:1]
    disease_vecs = tfidf[1:]
    sims = cosine_similarity(conv_vec, disease_vecs).flatten()
    return sims

# ---------- 6. PROCESS EACH CONVERSATION & PREDICT DISEASE ----------
all_rows = []

for idx, conv in enumerate(conversations, start=1):
    conv_id = f"Conversation_{idx}"
    print("====================================")
    print(conv_id)
    print("====================================\n")

    # 1) Extract age and symptoms from this conversation
    p_age = extract_age_from_text(conv)
    p_symptoms = extract_symptoms_from_text(conv, symptom_keywords)

    if p_symptoms:
        conv_symptom_text = " ; ".join(p_symptoms)
    else:
        print("⚠️ No explicit symptoms (from CSV list) detected. Using full conversation text.")
        conv_symptom_text = conv

    # 2) Prepare disease text (symptoms + comorbidities)
    disease_texts = (
        disease_df["symptoms"].fillna("").astype(str)
        + " ; "
        + disease_df["related_comorbidities"].fillna("").astype(str)
    ).tolist()

    # 3) Lexical similarity (TF-IDF + cosine)
    lexical_scores = compute_lexical_similarity(conv_symptom_text, disease_texts)

    # 4) Age score for each disease
    age_scores = []
    for _, row in disease_df.iterrows():
        a_s = age_score(p_age, row["parsed_age_range"])
        age_scores.append(a_s)

    # 5) Final score (symptoms + age)
    SYMPTOM_WEIGHT = 0.7
    AGE_WEIGHT = 0.3

    tmp = disease_df.copy()
    tmp["lexical_similarity_score"] = lexical_scores
    tmp["age_score"] = age_scores
    tmp["final_score"] = (
        SYMPTOM_WEIGHT * tmp["lexical_similarity_score"]
        + AGE_WEIGHT * tmp["age_score"]
    )

    tmp_sorted = tmp.sort_values(by="final_score", ascending=False)

    # Print quick summary for this conversation
    print(f"Detected patient age: {p_age}")
    print("Extracted symptoms:")
    for s in p_symptoms:
        print("  -", s)
    print("\nTop 3 predicted diseases for this conversation:")
    for _, row in tmp_sorted.head(3).iterrows():
        print(f"- {row['disease']}")
        print(f"  final_score: {row['final_score']:.4f}")
        print(f"  lexical_similarity_score: {row['lexical_similarity_score']:.4f}")
        print(f"  age_score: {row['age_score']:.2f}")
        print(f"  typical_age_group: {row['typical_age_group']}")
        print(f"  disease_symptoms: {row['symptoms']}")
        print(f"  disease_comorbidities: {row['related_comorbidities']}\n")

    # Store top 5 for this conversation in combined results
    for _, row in tmp_sorted.head(5).iterrows():
        all_rows.append({
            "conversation_id": conv_id,
            "patient_age": p_age,
            "patient_symptoms": "; ".join(p_symptoms),
            "disease": row["disease"],
            "final_score": row["final_score"],
            "lexical_similarity_score": row["lexical_similarity_score"],
            "age_score": row["age_score"],
            "typical_age_group": row["typical_age_group"],
            "disease_symptoms": row["symptoms"],
            "disease_comorbidities": row["related_comorbidities"],
        })

# ---------- 7. SAVE RESULTS FOR ALL CONVERSATIONS ----------
if all_rows:
    result_df = pd.DataFrame(all_rows)
    result_df.to_csv(OUTPUT_FILE, index=False, encoding="utf-8")
    print(f"\n✅ All predictions saved to: {OUTPUT_FILE}")
else:
    print("\n⚠️ No predictions generated.")
```

```python
# ============================================================
# Predict diseases for multiple conversations using:
# - Symptoms (60%)
# - Age (25%)
# - Comorbidities (15%)
#
# Input:
#   - "generated_conversations (1).txt"  (multi-convo file)
#   - "prostate_diseases_symptoms_age_comorbidities.csv"
#
# Output:
#   - "generated_conversations_predictions.csv"
# ============================================================

import re
from pathlib import Path

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ---------- FILE PATHS ----------
CONV_PATH = Path("generated_conversations (1).txt")
CSV_PATH  = Path("prostate_diseases_symptoms_age_comorbidities.csv")
OUT_CSV   = Path("generated_conversations_predictions.csv")

# ---------- SAFETY CHECKS ----------
if not CONV_PATH.exists():
    raise FileNotFoundError(f"Conversation file not found: {CONV_PATH}")

if not CSV_PATH.exists():
    raise FileNotFoundError(f"Disease CSV file not found: {CSV_PATH}")

# ---------- 1. LOAD DISEASE DATA ----------
disease_df = pd.read_csv(CSV_PATH)

required_cols = ["disease", "symptoms", "typical_age_group", "related_comorbidities"]
for col in required_cols:
    if col not in disease_df.columns:
        raise ValueError(
            f"Column '{col}' is missing in the CSV. "
            f"Columns present: {list(disease_df.columns)}"
        )

# ---------- 2. AGE RANGE PARSING FOR DATASET ----------
def parse_age_range(age_text):
    """
    Examples in CSV:
      'Usually men >50 (common by 60-80)'
      'Younger to middle-aged men (20-50)'
      'Older men'
    We try to extract numbers and build (low, high) ranges.
    """
    if not isinstance(age_text, str):
        age_text = str(age_text)
    t = age_text.lower().strip()
    nums = re.findall(r'(\d{1,3})', t)

    # e.g. ">50", "over 60"
    if ("above" in t or "over" in t or ">" in t) and nums:
        low = int(nums[0])
        return (low, 120)

    # e.g. "below 40"
    if ("below" in t or "under" in t or "<" in t) and nums:
        high = int(nums[0])
        return (0, high)

    # e.g. "20-50", "20 to 50"
    if len(nums) >= 2:
        low, high = int(nums[0]), int(nums[1])
        if low > high:
            low, high = high, low
        return (low, high)

    # e.g. single "60" → treat as 60 ± 10
    if len(nums) == 1:
        v = int(nums[0])
        return (max(0, v - 10), v + 10)

    # Things like "Older men" / "Adults" → unknown numeric band
    return None

disease_df["parsed_age_range"] = disease_df["typical_age_group"].apply(parse_age_range)

# ---------- 3. KEYWORD LISTS FROM CSV (for extraction) ----------
def build_keywords(df, col):
    all_items = []
    for s in df[col].fillna(""):
        parts = re.split(r"[;,]", str(s))
        for p in parts:
            p = p.strip().lower()
            if p:
                all_items.append(p)
    seen = set()
    out = []
    for x in all_items:
        if x not in seen:
            seen.add(x)
            out.append(x)
    # longer phrases first
    return sorted(out, key=len, reverse=True)

symptom_keywords     = build_keywords(disease_df, "symptoms")
comorbidity_keywords = build_keywords(disease_df, "related_comorbidities")

# ---------- 4. LOAD & SPLIT CONVERSATIONS FROM TXT ----------
full_text = CONV_PATH.read_text(encoding="utf-8")

def split_conversations(text):
    """
    Expects blocks starting like:
      Conversation 1:
      Conversation 2:
    separated by lines of '---------'
    """
    lines = text.splitlines()
    conversations = []
    current_lines = []
    current_id = None

    for line in lines:
        stripped = line.strip()
        if stripped.startswith("Conversation ") and stripped.endswith(":"):
            # flush previous conv
            if current_lines and current_id is not None:
                conv_text = "\n".join(current_lines).strip()
                conversations.append((current_id, conv_text))
                current_lines = []
            # new conv id
            # e.g. "Conversation 1:" -> "Conversation_1"
            num_match = re.search(r"Conversation\s+(\d+)", stripped, re.IGNORECASE)
            if num_match:
                conv_num = num_match.group(1)
                current_id = f"Conversation_{conv_num}"
            else:
                current_id = stripped.replace(" ", "_").rstrip(":")
            continue

        # ignore separator lines
        if stripped.startswith("---------"):
            continue

        # regular content line
        if stripped != "" or current_lines:
            current_lines.append(line)

    # last block
    if current_lines and current_id is not None:
        conv_text = "\n".join(current_lines).strip()
        conversations.append((current_id, conv_text))

    # If nothing recognized, treat whole text as one conversation
    if not conversations:
        conversations = [("Conversation_1", text.strip())]

    return conversations

conversations = split_conversations(full_text)
print(f"🔍 Found {len(conversations)} conversation(s) in the file.\n")

# ---------- 5. HELPERS FOR PER-CONVERSATION EXTRACTION ----------

def strip_diagnosis_line(text):
    """
    Remove lines like:
    'Doctor: Based on your symptoms... it appears you may be experiencing ...'
    so we don't 'cheat' by reading the diagnosis text itself.
    """
    lines = text.splitlines()
    kept = []
    for ln in lines:
        if "appears you may be experiencing" in ln.lower():
            continue
        kept.append(ln)
    return "\n".join(kept)

def extract_age_from_text(text):
    """
    For this synthetic file, age is given as:
      'I am from the age group Younger to middle-aged men (20-50).'
    We'll:
      - look for the 'age group' line
      - extract numbers from that line
      - if 2+ numbers, use their average
    """
    t = text.lower()
    age_line = None

    for line in text.splitlines():
        if "age group" in line.lower():
            age_line = line
            break

    if age_line:
        nums = re.findall(r'(\d{1,3})', age_line)
        if len(nums) >= 2:
            vals = [int(n) for n in nums]
            return int(sum(vals) / len(vals))
        elif len(nums) == 1:
            return int(nums[0])

    # fallback: try patterns like "45 years old"
    patterns = [
        r"(\d{1,3})\s*(?:years? old|yrs? old|y/o|years)",
        r"\b(\d{1,3})-year-old\b",
        r"age\s*(?:is|:)?\s*(\d{1,3})",
        r"\bi am\s+(\d{1,3})\b",
    ]
    for p in patterns:
        m = re.search(p, t)
        if m:
            try:
                return int(m.group(1))
            except:
                pass
    return None

def extract_keywords_from_text(text, keywords):
    t = text.lower()
    found = []
    for kw in keywords:
        pattern = r"\b" + re.escape(kw) + r"\b"
        if re.search(pattern, t):
            found.append(kw)
    return sorted(set(found))

# ---------- 6. SCORE FUNCTIONS ----------
def age_score(patient_age, age_range):
    if patient_age is None or age_range is None:
        return 0.0
    low, high = age_range
    if low <= patient_age <= high:
        return 1.0
    if (patient_age >= low - 10) and (patient_age <= high + 10):
        return 0.5
    return 0.0

def comorbidity_score(patient_comorbs, disease_comorbs_str):
    if not patient_comorbs:
        return 0.0
    disease_comorbs = [
        c.strip().lower()
        for c in re.split(r"[;,]", str(disease_comorbs_str))
        if c.strip()
    ]
    if not disease_comorbs:
        return 0.0
    pset = set(patient_comorbs)
    dset = set(disease_comorbs)
    inter = pset.intersection(dset)
    return len(inter) / max(1, len(dset))

def compute_lexical_similarity(conv_text, disease_texts):
    """
    TF-IDF + cosine similarity between:
    - conv_text (cleaned conversation)
    - each disease's combined symptoms+comorbidities
    """
    texts = [conv_text] + disease_texts
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf = vectorizer.fit_transform(texts)
    conv_vec = tfidf[0:1]
    disease_vecs = tfidf[1:]
    sims = cosine_similarity(conv_vec, disease_vecs).flatten()
    return sims

# Precompute combined text for each disease (symptoms + comorb)
disease_df["combined_text"] = (
    disease_df["symptoms"].fillna("").astype(str)
    + " ; "
    + disease_df["related_comorbidities"].fillna("").astype(str)
)

# ---------- 7. PROCESS EACH CONVERSATION ----------
all_rows = []

for conv_id, conv_text in conversations:
    print("====================================")
    print(conv_id)
    print("====================================\n")

    # Remove the line that already states the disease diagnosis
    conv_clean = strip_diagnosis_line(conv_text)

    # Extract features
    p_age   = extract_age_from_text(conv_clean)
    p_syms  = extract_keywords_from_text(conv_clean, symptom_keywords)
    p_comor = extract_keywords_from_text(conv_clean, comorbidity_keywords)

    # For similarity, we use the cleaned full conversation text
    conv_for_similarity = conv_clean

    # Lexical (symptom) similarity
    symptom_scores = compute_lexical_similarity(
        conv_for_similarity,
        disease_df["combined_text"].tolist()
    )

    # Age & comorbidity scores
    age_scores_list = []
    comorb_scores_list = []

    for _, row in disease_df.iterrows():
        a_s = age_score(p_age, row["parsed_age_range"])
        c_s = comorbidity_score(p_comor, row["related_comorbidities"])
        age_scores_list.append(a_s)
        comorb_scores_list.append(c_s)

    tmp = disease_df.copy()
    tmp["symptom_score"]      = symptom_scores
    tmp["age_score"]          = age_scores_list
    tmp["comorbidity_score"]  = comorb_scores_list

    # Your EXACT formula:
    W_SYMPTOM = 0.60   # 60%
    W_AGE     = 0.25   # 25%
    W_COMORB  = 0.15   # 15%

    tmp["final_score"] = (
        W_SYMPTOM * tmp["symptom_score"]
        + W_AGE   * tmp["age_score"]
        + W_COMORB* tmp["comorbidity_score"]
    )

    tmp_sorted = tmp.sort_values(by="final_score", ascending=False)

    # Show quick summary
    print(f"Detected patient age (approx): {p_age}")
    print("Extracted symptoms:")
    for s in p_syms:
        print("  -", s)
    print("Extracted comorbidities:")
    for c in p_comor:
        print("  -", c)

    print("\nTop 3 predicted diseases with final_score (60/25/15):")
    for _, row in tmp_sorted.head(3).iterrows():
        print(f"- {row['disease']}")
        print(f"  final_score: {row['final_score']:.4f}")
        print(f"  symptom_score: {row['symptom_score']:.4f}")
        print(f"  age_score: {row['age_score']:.2f}")
        print(f"  comorbidity_score: {row['comorbidity_score']:.2f}")
        print(f"  typical_age_group: {row['typical_age_group']}")
        print(f"  disease_symptoms: {row['symptoms']}")
        print(f"  disease_comorbidities: {row['related_comorbidities']}\n")

    # Save top 5 for this conversation into combined results
    for _, row in tmp_sorted.head(5).iterrows():
        all_rows.append({
            "conversation_id": conv_id,
            "patient_age_estimated": p_age,
            "patient_symptoms": "; ".join(p_syms),
            "patient_comorbidities": "; ".join(p_comor),
            "disease": row["disease"],
            "final_score": row["final_score"],
            "symptom_score": row["symptom_score"],
            "age_score": row["age_score"],
            "comorbidity_score": row["comorbidity_score"],
            "typical_age_group": row["typical_age_group"],
            "disease_symptoms": row["symptoms"],
            "disease_comorbidities": row["related_comorbidities"],
        })

# ---------- 8. SAVE ALL PREDICTIONS ----------
if all_rows:
    result_df = pd.DataFrame(all_rows)
    result_df.to_csv(OUT_CSV, index=False, encoding="utf-8")
    print(f"\n✅ All predictions saved to: {OUT_CSV}")
else:
    print("\n⚠️ No predictions generated.")
```
